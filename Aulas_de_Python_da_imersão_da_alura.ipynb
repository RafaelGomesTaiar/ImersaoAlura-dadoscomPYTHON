```python
df.columns
```

```python
df.rename(columns={
    'ano':'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

display(df.head())
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

display(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Media',
    'L': 'Grande'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    100: 'Totalmente Remoto',
    50: 'Hbrido',
    0: 'Presencial'
})

display(df['remoto'].value_counts())
```

```python
import numpy as np
#criacao de um dataframe de teste para usar como exemplo
df_salarios = pd.DataFrame({
    'nomes': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})
#calcula a mdia e substitui os nulos e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#calcula mediana e substitui os nuloas pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dias': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dias': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nomes': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade':['So Paulo', np.nan, 'Curitiba', np.nan, 'Belm']
    })

df_cidades ['cidade_preenchida'] = df_cidades['cidade'].fillna('No informado')
display(df_cidades)
```

# AULA 3 DE PYTHON - CRIE GRFICOS E CONTE HISTRIAS COM DADOS


```python
df_limpo.head()
```

```python
df_limpo['senioridade'].value_counts().plot(kind='bar', title='distribuicao de senioridade')
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title('salario mdio por nvel de senioridade')
plt.xlabel('senioridade')
plt.ylabel('salario medio anual (usd)')
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title('salario mdio por nvel de senioridade')
plt.xlabel('senioridade')
plt.ylabel('salario medio anual (usd)')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df_limpo['usd'], bins = 250, kde =False)
plt.title('distribuicao de salarios anuais')
plt.xlabel('salario em (usd)')
plt.ylabel('frequncia')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['usd'])
plt.title('box plot salrio')
plt.xlabel('salario em (usd)')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title('boxplot da distribuicao por senioridade')
plt.xlabel('salario em (usd)')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title('boxplot da distribuicao por senioridade')
plt.xlabel('salario em (usd)')
plt.show()
```

```python
import plotly.express as px
```

```python
df_senioridade_ordenado = df_limpo.groupby('senioridade')['usd'].mean().reset_index().sort_values(by='usd', ascending=False)

fig = px.bar(df_senioridade_ordenado,
             x='senioridade',
             y='usd',
             title='Mdia salarial por nvel de senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Salrio Mdio Anual (USD)'})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']


fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='proporo dos tipos de trabalho',
             hole=0.5)
fig.update_traces(textinfo='percent+label')
fig.show()
```
